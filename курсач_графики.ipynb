```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import seaborn as sns  
import pandas as pd
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.integrate import odeint

y0 = [1]
y1 = [5000]
y2 = [20000]
t = np.linspace(0,500,num=1000)
r = 0.05
K = 10000

params = [r,K]

def sim(variables,t,params):
    X = variables[0]
    r = params[0]
    K = params[1]
    dXdt = r*X * (1 - X/K)
    return([dXdt])

y = odeint(sim,y0, t, args=(params,))
y_1 = odeint(sim,y1, t, args=(params,))
y_2 = odeint(sim,y2, t, args=(params,))
```

```python
var11  = np.arange(start=0, stop=1000, step=1).reshape(-1)
var12 = y[:,0]

dataset = pd.DataFrame()
dataset['timepoint'] = var11
dataset['cell number, N0 < r'] = var12

var22 = y_1[:,0]

dataset['cell number, N0 = 1/2r'] = var22


var32 = y_2[:,0]

dataset['cell number, N0 = r'] = var32
```

```python
dataset = dataset.drop(['timepoint'], axis = 1)
dataset
```

```python
plt.figure(figsize=(7,5))

plt.rcParams['figure.dpi'] = 1000

sns.set_palette("husl", 3)

p = sns.lineplot(data = dataset)
p.set_xlabel("Время", fontsize = 10)
p.set_ylabel("Численность популяции", fontsize = 10)
p.set_title("Логистический рост при различных начальных значениях численности популяции") 



plt.savefig("logistic_growth.png",bbox_inches ="tight",
            pad_inches = 1,
            transparent = True,
            facecolor ="w",
            edgecolor ='w',
            orientation ='landscape')
```

```python
def odeEuler(f,y0,t):
    y = np.zeros(len(t))
    y[0] = y0
    for n in range(0,len(t)-1):
        y[n+1] = y[n] + f(y[n],t[n])*(t[n+1] - t[n])
    return y
```

```python
t = np.linspace(0,2,5)
t1 = np.linspace(0,2,10)


y0 = 1
f = lambda y,t: y

y_true = np.exp(t)
y_1 = odeEuler(f,y0,t)
y_05 = odeEuler(f,y0,t1)

```

```python
df = pd.DataFrame()
df['Значение функции'] = y_true
df['Аппроксимация методом Эйлера, h = 1'] = y_1
```

```python
plt.figure(figsize=(7,5))
plt.rcParams['figure.dpi'] = 1500

sns.set_palette("husl", 3)

p = sns.lineplot(data = df, marker = 'o')
p.set_xlabel("Время", fontsize = 10)
p.set_ylabel("Значение функции", fontsize = 10)
p.set_title("Аппроксимация функции методом Эйлера") 
```
